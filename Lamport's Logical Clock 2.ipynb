```python
class LamportClock:
    def __init__(self):
        self.time = 0

    def increment(self):
        self.time += 1

    def update(self, received_time):
        self.time = max(self.time, received_time) + 1

    def get_time(self):
        return self.time


# Create Lamport clocks for two processes
clock_A = LamportClock()
clock_B = LamportClock()

def display_time():
    print(f"Process A time: {clock_A.get_time()}")
    print(f"Process B time: {clock_B.get_time()}")

while True:
    print("\nChoose an action:")
    print("1: Internal event in Process A")
    print("2: Internal event in Process B")
    print("3: Send message from Process A to Process B")
    print("4: Send message from Process B to Process A")
    print("5: Display current times")
    print("6: Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        # Internal event in Process A
        clock_A.increment()
        print("Internal event in Process A occurred.")
    elif choice == 2:
        # Internal event in Process B
        clock_B.increment()
        print("Internal event in Process B occurred.")
    elif choice == 3:
        # Message from Process A to Process B
        message_time = clock_A.get_time()
        clock_B.update(message_time)
        print("Message sent from Process A to Process B.")
    elif choice == 4:
        # Message from Process B to Process A
        message_time = clock_B.get_time()
        clock_A.update(message_time)
        print("Message sent from Process B to Process A.")
    elif choice == 5:
        # Display current times
        display_time()
    elif choice == 6:
        print("Exiting simulation.")
        break
    else:
        print("Invalid choice, please try again.")

    # Display updated times after each action
    display_time()
```
